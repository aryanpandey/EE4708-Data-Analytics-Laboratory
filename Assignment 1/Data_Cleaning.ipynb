```python
print(len(merged_data['State'].unique()))

merged_data = merged_data[merged_data['State']!='NV']
merged_data = merged_data[merged_data['State']!='KS']
merged_data = merged_data[merged_data['State']!='MN']

len(merged_data['State'].unique())
```

```python
cols_to_fill = ['Med_Income', 'Med_Income_White', 'Med_Income_Black',
       'Med_Income_Nat_Am', 'Med_Income_Asian', 'Hispanic']

def fill_na_with_mean_median_income(df):
    for col in cols_to_fill:
        col_mean = np.mean(df[col])
        df[col].fillna(col_mean, inplace = True)
    return df

filled_merged_data = pd.DataFrame()

for grp in merged_data['State'].unique():
    df = merged_data[merged_data['State'] == grp]
    
    df = fill_na_with_mean_median_income(df)
    
    filled_merged_data = filled_merged_data.append(df)

filled_merged_data.isna().sum()/29.25
```

```python
filled_merged_data.info()
```

```python
cols_with_star = ['Incidence_Rate', 'recent_trend', 'Mortality_Rate', 'Avg_Ann_Deaths']

len(filled_merged_data[filled_merged_data['Incidence_Rate'] == '*'])*100/len(filled_merged_data)
```

```python
for col in cols_with_star:
    filled_merged_data = filled_merged_data[filled_merged_data[col] != '*']
    
len(filled_merged_data)
```

```python
def fix_hashed_strings(row):
    try:
        row['Incidence_Rate'] = float(row['Incidence_Rate'].replace(' #', ''))
    except:
        pass
    
    return row

filled_merged_data = filled_merged_data.apply(fix_hashed_strings, axis = 1)
filled_merged_data['Incidence_Rate'].unique()
```

```python
filled_merged_data.info()
```

```python
filled_merged_data.to_csv('clean_merged.csv', index = False)
```

```python
cat_cols = ['Age-Adjusted Incidence Rate(Ê) - cases per 100,000',
       'Lower 95% Confidence Interval', 'Upper 95% Confidence Interval',
       'Average Annual Count', 'Recent Trend',
       'Recent 5-Year Trend (ˆ) in Incidence Rates',
       'Lower 95% Confidence Interval.1', 'Upper 95% Confidence Interval.1']
    
incd[cat_cols].apply(get_unique_vals)
```

```python
weird_characters = ['*', '_', '__', '__ ', '_ ', '* ', ' _', ' *', ' __', '3 or fewer', '']

def fix_strings(row):
    try:
        row['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'] = float(row['Age-Adjusted Incidence Rate(Ê) - cases per 100,000'].replace(' #  ', ''))
    except:
        pass
    
    for col in cat_cols:
        if row[col] in weird_characters:
            row[col] = np.nan
    
    return row

incd = incd.apply(fix_strings, axis = 1)
incd.info()
```

```python
incd_clean = incd.dropna()
```

```python
incd_clean.to_csv('incd_clean.csv', index = False)
```

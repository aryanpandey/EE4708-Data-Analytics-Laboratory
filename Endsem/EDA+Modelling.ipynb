```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cognizant = pd.read_csv('Datasets/Cognizant share prices 2019_2021.csv')
hcl = pd.read_csv('Datasets/HCL Technologies share prices 2019_2021.csv')
hdfc = pd.read_csv('Datasets/HDFC Bank Share Prices 2019_2021.csv')
icici = pd.read_csv('Datasets/ICICI Bank Share Prices 2019_2021.csv')
infosys = pd.read_csv('Datasets/Infosys Share Prices 2019_2021.csv')
sbi = pd.read_csv('Datasets/SBI Share Prices 2019_2021.csv')
usdinr = pd.read_csv('Datasets/USD-INR Exchange rate 2019_2021.csv')
```

```python
df_list = [cognizant, hcl, hdfc, icici, infosys, sbi, usdinr]
```

```python
for df in df_list:
    print(df.shape)
```

```python
for j, df in enumerate(df_list):
    changes = []
    for i in range(len(df)):
        if i==0:
            changes.append((df['Close'][i] - df['Open'][i])*100/df['Open'][i])
        else:
            changes.append((df['Close'][i] - df['Close'][i-1])*100/df['Close'][i-1])
    df['Percentage Change'] = changes
    df_list[j] = df
```

```python
title_list = ['Cognizant', 'HCL Tech', 'HDFC Bank', 'ICICI Bank', 'Infosys', 'SBI', 'USD-INR']
for i, df in enumerate(df_list):
    plt.figure(figsize = (8,8))
    df['Date'] = pd.to_datetime(df['Date'])
    plt.plot(df['Date'], df['Open'])
    plt.title(title_list[i] + ' Opening Price', size = 22)
    plt.xlabel('Date', size = 18)
    plt.ylabel('Share Value', size = 18)
    plt.xticks(rotation = 90, size = 16)
    plt.yticks(size = 16)
    
plt.show()
```

```python
for i, df in enumerate(df_list):
    if i!=6:
        plt.figure(figsize = (8,8))
        df['Date'] = pd.to_datetime(df['Date'])
        plt.plot(df['Date'], df['Volume'])
        plt.title(title_list[i] + ' Volume', size = 22)
        plt.xlabel('Date', size = 18)
        plt.ylabel('Volume', size = 18)
        plt.xticks(rotation = 90, size = 16)
        plt.yticks(size = 16)
    
plt.show()
```

```python
for i in range(len(df_list)):
    plt.figure(figsize = (8,8))
    plt.hist(df_list[i]['Percentage Change'], bins = 80)
plt.show()
```
